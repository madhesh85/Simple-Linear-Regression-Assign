```python
import pandas as pd
```

```python
load the dataset
```

```python
dataset=pd.read_csv("Salary_Data.csv")
```

```python
split the input and output
```

```python
independent=dataset[["YearsExperience"]]
```

```python
dependent=dataset[["Salary"]]
```

```python
split trainset and testset
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.3,random_state=0)
```

```python
slope and bias finding
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
weight=regressor.coef_
weight
```

```python
bias=regressor.intercept_
bias
```

```python
model creation
```

```python
y_pred=regressor.predict(x_test)
```

```python
Evaluation metrics - near to 1 then Good Model, Near to 0 thenn Bad model ,create one more
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
near to 1 so, its a good model
```

```python
save the model
```

```python
import pickle
```

```python
filename="finalised_model_linear.sav"
pickle.dump(regressor,open(filename,"wb"))
loaded_model=pickle.load(open("finalised_model_linear.sav","rb"))
result=loaded_model.predict([[15]])
```
